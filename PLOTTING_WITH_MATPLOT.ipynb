# Plotting Simple Curves

## Import and Settings

We will import `NumPy` and `matplotlib`. In addition, we will also start with some customised layout for the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Partitioning the real line

In order to plot a curve, we need a set of $x$ values and the corresponding $y$ values. Since $x$ is the independent variable, we need to first generate a list of values for $x$.

In `NumPy`:

```python
x = np.linspace(0,1,10) # this partitions the interval (0 , 1) in 10 points.
print(x.shape)
print(x)
```

## Curves

We shall plot several curves, starting from simple curvies to more complex ones. In this process, we will also learn how to add simple annotations to the plot.

### Curve-1

Plot $y = x$ for $x \in [0, 1]$. Sample $10$ equally spaced points in this interval, inclusive of endpoints.

In `NumPy`:

```python
y = x
plt.plot(x,y) # here order matters , please pass x-axis element 1st , then the y-axis elements.
```

```python
plt.scatter(x,y) # Scatter-plot
```

### Curve-2

Plot $y = 5 - 3x$ for $x \in [-5, 5]$. Sample $20$ equally spaced points in this interval, inclusive of endpoints. Add a title to the plot which has the equation of the curve.

```python
x = np.linspace(-5,5,20)
y = 5-3*x
plt.plot(x,y)
plt.title('y=5-3*x'); # this ';' sig is used to remove some extra elements in the title . U can remove it and check it , how it looks.
```

### Curve-3

Plot $y = x^2$ for $x \in [-1, 1]$. Try out four different samples for `x`:

- $5$ equally spaced points in this interval, endpoints inclusive
- $10$ equally spaced points in this interval, endpoints inclusive
- $20$ equally spaced points in this interval, endpoints inclusive
- $50$ equally spaced points in this interval, endpoints inclusive

Observe the differences in these four plots. Which one would you choose? Add the equation of the curve as the title of the plot. Also add the x and y-axis to the plot.

```python
x = np.linspace(-1,1,5)
y = x**2
plt.plot(x,y); # this graph is collection of some discrete points , so it is not smooth parabola
# using ';' after plt.plot removes the unwanted title from the graph heading
```

```python
x = np.linspace(-1,1,1000)
y = x**2
plt.plot(x,y); # we increased partitions , which in turn resulted in smooth parabola
```

### Curve-4

Plot $y = 3$ and $x = 5$ on the same plot. Color the first one green and the second one red. Limit your plot to the region $x \in [0, 6]$ and $y \in [0, 6]$.

- Label the x and the y axes.
- Add a suitable title.
- Add a legend.

```python
x = np.linspace(0,6) # by default this creates 50 point partitions.
y = np.ones(50)*3
plt.plot(x,y , color = 'red' , label = 'y = 3');

x = np.ones(50)*5
y = np.linspace(0,6)
plt.plot(x , y , color = 'green' , label = 'x = 5');

plt.legend() # this activates the label views , which is visible in the graph.
plt.title('Y = 3 and X = 5');
```

## Subplots

Sometimes we may have to plot multiple objects, each in a separate plot. Plot the following curves, each in a separe plot:

- $y = \ln x, \quad x \in \left( \cfrac{1}{e^2}, e^3 \right)$

- $y = e^x, \quad x \in (0, 3)$

- $y = \sin x, \quad x \in [0, 2 \pi]$

- $y = \cfrac{1}{x}, \quad x \in (0, 5)$


Title each curve appropriately.

```python
plt.rcParams['figure.figsize'] = [9,9] # rcParmas is a extention of matplotlib which will help to configure the fonts , sizes and other visuals in these subplots.

#plot - 1
plt.subplot(2,2,1) # plt.subplot(a,b,c) means area is divided into a rows and b columns and then c is the position where u will be ploting a particular figure.
x = np.linspace(1/np.e**2 , np.e**3)
plt.plot(x , np.log(x))
plt.title('y = lnx');

#plot - 2
plt.subplot(2,2,2)
x = np.linspace(0,3)
plt.plot(x , np.exp(x))
plt.title('$y = e^x$');

#plot - 3
plt.subplot(2,2,3)
x = np.linspace(0 , np.pi)
plt.plot(x , np.sin(x))
plt.title('y = sin(x)');

#plot - 4
plt.subplot(2,2,4)
x = np.linspace(0,5)
plt.plot(x , 1/x)
plt.title('$y= 1/x$');

plt.savefig('fig1.png') #we can save this image in instances by this code
```
